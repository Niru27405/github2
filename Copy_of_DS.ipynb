```python
import pandas as pd
from google.colab import files

# Step 1: Upload the CSV
uploaded = files.upload()

# Get the filename of the uploaded CSV
filename = list(uploaded.keys())[0]

# Step 2: Read the CSV into a DataFrame
df = pd.read_csv(filename)

# Check the column names to ensure they are as expected
print("Columns in the DataFrame:", df.columns)

# Step 3: Add a new column 'Is Draw' to check if the 'Winner' column is 'drawn'
df['Is Draw'] = df['Winner'] == 'drawn'

# Verify if the 'Is Draw' column was added correctly
print("First few rows of the DataFrame:")
print(df.head())

# Step 4: Perform the groupby operation
result = df.groupby('Ground').agg(
    Total_Matches=('Ground', 'count'),
    Total_Draws=('Is Draw', 'sum')
)

# Display the result
print(result)
```

```python
import pandas as pd
from google.colab import files

# Step 1: Upload the CSV
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Step 2: Read the CSV into a DataFrame
df = pd.read_csv(filename)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Winner', order=df['Winner'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title('Top 10 Winning Teams')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload the CSV
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Step 2: Read the CSV into a DataFrame
df = pd.read_csv(filename)

# Step 3: Count appearances of each team
Nteam_counts = pd.concat([df['Team 1'], df['Team 2']]).value_counts().head(10)

# Step 4: Plotting
plt.figure(figsize=(10, 5))
sns.barplot(x=Nteam_counts.index, y=Nteam_counts.values)
plt.xticks(rotation=45)
plt.title('Top 10 Most Frequent Teams')
plt.ylabel('Number of Matches')
plt.xlabel('Teams')
plt.tight_layout()
plt.show()
```
